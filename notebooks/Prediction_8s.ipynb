# Prediction of bone fractures using the trained models

## Prediction (Segmentation) using YOLO8s-seg model.

```python
!yolo task=segment mode=predict save=True model=WeightFiles/best_segmentation.pt conf=0.25 source="datasets/dataset/Pred"
```

```python
from PIL import Image
import os
import random

folder_path = 'runs/segment/predict'


image_filenames = os.listdir(folder_path)
random_images = random.sample(image_filenames, 4)
grid_image = Image.new('RGB', (800, 800))


for i in range(4):

    current_image = Image.open(os.path.join(folder_path, random_images[i]))
    current_image.thumbnail((400, 400))
    x = (i % 2) * 400
    y = (i // 2) * 400
    grid_image.paste(current_image, (x, y))

display(grid_image)
```
